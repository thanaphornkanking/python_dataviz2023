````python
def grading(x):
    if type(x) is int:     # type(x) ตรวจสอบชนิดของตัวแปร
        if x < 0:
            y =f'คะแนน {x} ไม่สามารถตัดเกรดได้'
        elif x < 50:
            y = f'คะแนน {x} ได้เกรด F'
        elif x < 55:
            y = f'คะแนน {x} ได้เกรด D'
        elif x < 60:
            y = f'คะแนน {x} ได้เกรด D+'
        elif x < 65:
            y = f'คะแนน {x} ได้เกรด C'
        elif x < 70:
            y = f'คะแนน {x} ได้เกรด C+'
        elif x < 75:
            y = f'คะแนน {x} ได้เกรด B'
        elif x < 80:
            y = f'คะแนน {x} ได้เกรด B+'
        elif x <= 100:
            y = f'คะแนน {x} ได้เกรด A'
        else:
            y =f'คะแนน {x} ไม่สามารถตัดเกรดได้'
    else:
        y =f'คะแนน {x} ไม่สามารถตัดเกรดได้'

    return y
````

````python
sep_df = pd.read_excel('/content/drive/MyDrive/DataViz23Data/sep65.xlsx',header=8,skipfooter=26)
sep_df = sep_df.iloc[-2:,:15]
sep_df['เดือน-ปี'] = len(sep_df)*['sep-65']
sep_df
````

````python
oct_df = pd.read_excel('/content/drive/MyDrive/DataViz23Data/oct65.xlsx',header=9,skipfooter=27)
oct_df = oct_df.iloc[-2:,:15]
oct_df['เดือน-ปี'] = len(oct_df)*['oct-65']
oct_df
````

# ต่อตารางแกน Y (แกน rows เพิ่มจำนวนข้อมูล)

````python
sep_oct_df = pd.concat([sep_df,oct_df])
sep_oct_df
````

ข้อ 1 งบประมาณประจำหรืองบประมาณลงทุน สูงกว่ากัน

````python
##งบประจำ sep และงบลงทุน sep
a =sep_oct_df[(sep_oct_df['รายการ']=='ประจำ')&( sep_oct_df['เดือน-ปี']=='sep-65')]['งบประมาณ']
b =sep_oct_df[(sep_oct_df['รายการ']=='ลงทุน')&( sep_oct_df['เดือน-ปี']=='sep-65')]['งบประมาณ']
##งบประจำ sep และงบลงทุน oct
c =sep_oct_df[(sep_oct_df['รายการ']=='ประจำ')&( sep_oct_df['เดือน-ปี']=='oct-65')]['งบประมาณ']
d =sep_oct_df[(sep_oct_df['รายการ']=='ลงทุน')&( sep_oct_df['เดือน-ปี']=='oct-65')]['งบประมาณ']



if int(a)>int(b): print(f'ในเดือนกันยายน งบประมาณประจำ สูงกว่างบประมาณลงทุนอยู่',int(a)-int(b),'บาท')
else: print (f'ในเดือนกันยายน งบประมาณลงทุน สูงกว่างบประมาณประจำอยู่',int(b)-int(a),'บาท')

if int(c)>int(d): print(f'ในเดือนตุลาคม งบประมาณประจำ สูงกว่างบประมาณลงทุนอยู่', int(c)-int(d),'บาท')
else: print (f'ในเดือนตุลาคม งบประมาณลงทุน สูงกว่างบประมาณประจำอยู่', int(d)-int(c),'บาท')
````

ข้อ 2 งบรวม เดือน oct หรือ sep อันไหนมากกว่ากัน

````python
# ใช้ตัวแปร a b c d จากข้อที่ 1

S = int(a) + int(b) #sep
O = int(c) + int(d) #oct

if S > O: print(f'ฉะนั้นงบรวมเดือนกันยายนมากกว่าเดือนตุลาคมอยู่',(S-O),'บาท')
else: print(f'ฉะนั้นงบรวมเดือนตุลาคมมากกว่าเดือนกันยายนอยู่',(O-S),'บาท')
````

ข้อ 3 แก้ค่า missing พร้อมทั้งให้เหตุผล

````python
sep_oct_df = pd.concat([sep_df,oct_df])

sep_oct_df.rename(columns=lambda x: x.replace('\n', ''), inplace=True) #ลบค่า \n
sep_oct_df.rename(columns=lambda x: x.replace('*/-', ''), inplace=True) #ลบค่า */-
sep_oct_df.fillna('-',inplace=True) #ลบค่า NaN เป็น - เนื่องจากไม่มีข้อมูล

sep_oct_df
````

ข้อ 4 เดือนไหนเบิกจ่ายรวมมากกว่ากัน (การใช้จ่าย (PO+เบิกจ่าย))

````python

##การใช้จ่าย (PO+เบิกจ่าย) sep
s1 =sep_oct_df [( sep_oct_df['เดือน-ปี']=='sep-65')]['การใช้จ่าย (PO+เบิกจ่าย)'].sum()

##การใช้จ่าย (PO+เบิกจ่าย) oct
o1 =sep_oct_df[( sep_oct_df['เดือน-ปี']=='oct-65')]['การใช้จ่าย (PO+เบิกจ่าย)'].sum()


if int(s1) > int(o1): print(f'ฉะนั้นเดือนกันยายน เบิกจ่ายรวมมากกว่า เดือนตุลาคมอยู่',int(s1-o1),'บาท')
else: print(f'ฉะนั้นเดือนตุลาคม เบิกจ่ายรวมมากกว่า เดือนกันยายนอยู่',int(o1-s1),'บาท')


````
